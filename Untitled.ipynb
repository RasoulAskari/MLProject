# $$ x^2+x+1 $$

```python
def f(x):
    return x**2 + x + 1;

x_1 = np.linspace(start=-3, stop=3, num=100)
plt.xlabel("X")
plt.ylabel('Y')
plt.ylim(0.7, 12)
plt.plot(x_1, f(x_1))
plt.show()
```

```python
def df(x):
    return 2 * x + 1;
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.title("Cost Function", fontsize="17")
plt.xlabel("X", fontsize="16")
plt.ylabel("f(x)", fontsize="16")
plt.plot(x_1, f(x_1), color="blue", linewidth=3)
plt.xlim(-3, 3)
plt.ylim(0, 8)


plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1), color="skyblue", linewidth=5)
plt.title("Slope of Cost Function", fontsize="17")

plt.xlabel("X", fontsize="16" )
plt.ylabel("df(x)", fontsize="16")
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.show()
```

```python
new_x = -3
previos_x = 0
step_multiplier = 0.1
precision = 0.00001
x_list = [new_x]
slope_list = [df(new_x)]




for n in range(500):
    previos_x = new_x
    gradient = df(previos_x)
    new_x = previos_x - step_multiplier * gradient

    step_size = abs(new_x - previos_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))

    if step_size < precision:
        print("loop rans this many times: ", n)
        break
    
    
    
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.title("Cost Function", fontsize="17")
plt.xlabel("X", fontsize="16")
plt.ylabel("f(x)", fontsize="16")
plt.plot(x_1, f(x_1), color="blue", linewidth=3)
plt.xlim(-3, 3)
plt.ylim(0, 8)
values = np.array(x_list)
plt.scatter(x=x_list, y=f(values), s=100, alpha=0.6)




plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1), color="skyblue", linewidth=5, alpha=0.6)
plt.title("Slope of Cost Function", fontsize="17")
plt.xlabel("X", fontsize="16" )
plt.ylabel("df(x)", fontsize="16")
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.scatter(x=x_list, y=df(values), color="red", alpha=0.5, s=100)
plt.show()
```

```python
def gradient_descent(derivation_func, initial_guess, multiplier=0.02, perision=0.001, max_iter=300):
    new_x = initial_guess;
    x_list = [new_x]
    slope_list = [derivation_func(new_x)]
    
    
    for n in range(max_iter):
        previos_x = new_x
        gradient = derivation_func(previos_x)
        new_x = previos_x - multiplier * gradient
        
        step_size = abs(new_x - previos_x)
        x_list.append(new_x)
        slope_list.append(derivation_func(new_x))
        
        if step_size < precision:
            break
        
    return new_x, x_list, slope_list
        
    
    
```

## $${x, y}= \frac{1}{3^{-x^2 - y^2} + 1}$$
## $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2-y^2}$ 

```python
def f(x, y):
    r = 3**(-x**2 - y**2)
    return 1 / (r + 1)
```

```python
x_4 = np.linspace(start=-2, stop=2, num=200)
y_4 = np.linspace(start=-2, stop=2, num=200)


x_4, y_4 = np.meshgrid(x_4, y_4)

```

```python
fig = plt.figure(figsize= [16, 12])
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_4, y_4, f(x_4, y_4), cmap=cm.coolwarm, alpha=0.4)
plt.show()
```

```python
a, b = symbols('x, y')

print("Our Cost Function f(x, y) is: ", f(a, b))
print("Partial Derivative wrt x is: ", diff(f(a, b), a))
print("Value of x, y at a=1.8 and b=1.0 is: ", f(a, b).evalf(subs={a:1.8, b:1.0})) 
print("Value of Partial Dervative is: ", diff(f(a, b), a).evalf(subs={a: 1.8, b: 1.0}))
```

```python
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0])

for n in range(max_iter):
    gradient_x = diff(f(a, b), a).evalf(subs={a: params[0], b: params[1]})
    gradient_y = diff(f(a, b), b).evalf(subs={a: params[0], b: params[1]})
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier * gradients
    


    
print("Values in gradient array: ", gradients)
print("Minimum occurs at value of X of: ", gradient_x)
print("Minimum occurs at value of Y of: ", gradient_y)
print("Cost is: ", f(gradient_x, gradient_y))
```


## $$\frac {\partial f}{\partial x} = \frac {2x \ln(3) \cdot 3^{-x^2 - y^2}} {\left(3^{-x^2-y^2} +1 \right) ^ 2 }$$
## $$\frac {\partial f}{\partial y} = \frac {2y \ln(3) \cdot 3^{-x^2 - y^2}} {\left(3^{-x^2-y^2} +1 \right) ^ 2 }$$

```python
def fpx(x, y):
    r = 3**(-x**2 -y**2)
    return  2 * x * log(3) * r / (r + 1) ** 2
    
    
    

def fpy(x, y):
    r = 3**(-x**2 -y**2)
    return  2 * y * log(3) * r / (r + 1) ** 2

    
```

```python
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0])

for n in range(max_iter):
    gradient_x = fpx(params[0], params[1])
    gradient_y = fpy(params[0], params[1])
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier * gradients
    


    
print("Values in gradient array: ", gradients)
print("Minimum occurs at value of X of: ", gradient_x)
print("Minimum occurs at value of Y of: ", gradient_y)
print("Cost is: ", f(gradient_x, gradient_y))
```

## Example 5 Wroking with Data and Real Cost Function
## Mean Square Error: A Cost Function for Regression Problem
### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - \hat{y} \big)^2 $$


```python
x_5 = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose()
y_5 = np.array([1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]).reshape(7, 1)
```

```python
regr = LinearRegression()
regr.fit(x_5, y_5)

print("theta_0: ", regr.intercept_[0])
print("theta_1: ", regr.coef_[0][0])
```

```python
regr = LinearRegression()
regr.fit(x_5, y_5)





plt.scatter(x_5, y_5, s=50)
plt.plot(x_5, regr.predict(x_5), color="orange", linewidth=3)
plt.show()
```

```python
y_hat = 0.8475351486029545 + 1.2227264637835913 * x_5;

def mse(y, y_ha):
    mse_cale = (1 / y.size) * sum((y - y_ha) ** 2);
    return mse_cale

mse(y_5, y_hat)
```

## 3D Plot for the MSE Cost Function
### Make Data for Thetas

```python
nr_thetas = 200
th_0 = np.linspace(start=-1, stop=3, num=nr_thetas)
th_1 = np.linspace(start=-1, stop=3, num=nr_thetas)
plot_t0, plot_t1 = np.meshgrid(th_0, th_1)
```

### Calc MSE Using Nested for Loop

```python
plot_cost = np.zeros((nr_thetas, nr_thetas))

for i in range(nr_thetas):
    
    for j in range(nr_thetas):
        y_hat = plot_t0[i][j] + plot_t1[i][j] * x_5;
        plot_cost[i][j] = mse(y_5, y_hat)


```

```python
fig = plt.figure(figsize= [16, 12])
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(plot_t0, plot_t1, plot_cost,)
plt.show()
```
