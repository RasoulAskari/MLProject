# Example 1 - A Simple Cost Function

# $$ x^2+x+1 $$

```python
def f(x):
    return x**2 + x + 1;

x_1 = np.linspace(start=-3, stop=3, num=100)
plt.xlabel("X")
plt.ylabel('Y')
plt.ylim(0.7, 12)
plt.plot(x_1, f(x_1))
plt.show()
```

```python
def df(x):
    return 2 * x + 1;
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.title("Cost Function", fontsize="17")
plt.xlabel("X", fontsize="16")
plt.ylabel("f(x)", fontsize="16")
plt.plot(x_1, f(x_1), color="blue", linewidth=3)
plt.xlim(-3, 3)
plt.ylim(0, 8)


plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1), color="skyblue", linewidth=5)
plt.title("Slope of Cost Function", fontsize="17")

plt.xlabel("X", fontsize="16" )
plt.ylabel("df(x)", fontsize="16")
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.show()
```

```python
new_x = -3
previos_x = 0
step_multiplier = 0.1
precision = 0.00001
x_list = [new_x]
slope_list = [df(new_x)]




for n in range(500):
    previos_x = new_x
    gradient = df(previos_x)
    new_x = previos_x - step_multiplier * gradient

    step_size = abs(new_x - previos_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))

    if step_size < precision:
        print("loop rans this many times: ", n)
        break
    
    
    
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.title("Cost Function", fontsize="17")
plt.xlabel("X", fontsize="16")
plt.ylabel("f(x)", fontsize="16")
plt.plot(x_1, f(x_1), color="blue", linewidth=3)
plt.xlim(-3, 3)
plt.ylim(0, 8)
values = np.array(x_list)
plt.scatter(x=x_list, y=f(values), s=100, alpha=0.6)




plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1), color="skyblue", linewidth=5, alpha=0.6)
plt.title("Slope of Cost Function", fontsize="17")
plt.xlabel("X", fontsize="16" )
plt.ylabel("df(x)", fontsize="16")
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.scatter(x=x_list, y=df(values), color="red", alpha=0.5, s=100)
plt.show()
```

```python
x_2 = np.linspace(-2, 2, 1000)

def g(x):
    return x**4-4*x**2+5

def dg(x):
    return 4*x**3-8*x
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.plot(x_2, g(x_2))
plt.title("Slope of Cost Function", fontsize="17")
plt.xlabel("X", fontsize="16" )
plt.ylabel("df(x)", fontsize="16")
plt.xlim(-2, 2)




plt.subplot(1, 2, 2)
plt.plot(x_2, dg(x_2))
plt.title("Slope of Cost Function", fontsize="17")
plt.xlabel("X", fontsize="16" )
plt.ylabel("df(x)", fontsize="16")
plt.xlim(-2, 2)


plt.show() 
```

```python
def gradient_descent(derivation_func, initial_guess, multiplier=0.02, perision=0.001, max_iter=300):
    new_x = initial_guess;
    x_list = [new_x]
    slope_list = [derivation_func(new_x)]
    
    
    for n in range(max_iter):
        previos_x = new_x
        gradient = derivation_func(previos_x)
        new_x = previos_x - multiplier * gradient
        
        step_size = abs(new_x - previos_x)
        x_list.append(new_x)
        slope_list.append(derivation_func(new_x))
        
        if step_size < precision:
            break
        
    return new_x, x_list, slope_list
        
    
    
```

```python
local_min, list_x, deriv_list = gradient_descent(derivation_func=dg, initial_guess=-0.2)

```

```python
x_2 = np.linspace(-2, 2, 1000)

def g(x):
    return x**4-4*x**2+5

def dg(x):
    return 4*x**3-8*x
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.plot(x_2, g(x_2), color="blue", alpha=0.6)
plt.title("Slope of Cost Function", fontsize="17")
plt.xlabel("X", fontsize="16" )
plt.ylabel("df(x)", fontsize="16")
plt.xlim(-2, 2)
plt.scatter(x=list_x, y=g(np.array(list_x)),color="red" , alpha=0.5)



plt.subplot(1, 2, 2)
plt.plot(x_2, dg(x_2), linewidth=3, color="skyblue", alpha=0.6)
plt.title("Slope of Cost Function", fontsize="17")
plt.xlabel("X", fontsize="16" )
plt.ylabel("df(x)", fontsize="16")
plt.xlim(-2, 2)
plt.scatter(x=list_x, y=dg(np.array(list_x)),color="red" , alpha=0.5)


plt.show() 
```

## Example-3 Divergence, Overflow and Python Tuple
## $$ x^5+2x^4+2$$


```python
x_3 = np.linspace(start=-2.5, stop=2.5, num=1000 )


def h(x):
    return x**5-2*x**4+2

def dh(x):
    return 5*x**4-8*x**3


```

```python
local_min, list_x, deriv_list = gradient_descent(derivation_func=dh, initial_guess=-0.2, max_iter=71 )


plt.figure(figsize=[15, 5])



plt.subplot(1, 2, 1)
plt.plot(x_3, h(x_3), color="blue", linewidth=3, alpha=0.6)
plt.title("Divergence")
plt.xlabel("X", fontsize="16")
plt.ylabel("h(x)", fontsize="16")
plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)
plt.scatter(x=list_x, y=h(np.array(list_x)), color="red", alpha=0.5 , s=100)

plt.subplot(1, 2, 2)
plt.plot(x_3, dh(x_3),  color="skyblue", linewidth=3, alpha=0.6)
plt.xlabel("X", fontsize="16")
plt.ylabel("dh(x)", fontsize="16")
plt.xlim(-1, 2)
plt.ylim(-4, 5)
plt.scatter(x=deriv_list, y=dh(np.array(deriv_list)), color="red", alpha=0.5, s=100)

plt.show()
```
